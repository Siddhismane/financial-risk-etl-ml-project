# Create a Financial Risk Indicator

Loan Burden Ratio: Loan Burden Ratio

```python
# Create Loan Burden Feature
df["loan_to_income_ratio"] = df["loan_amount"] / df["annual_income"]
```

```python
# Preview
df[["loan_amount", "annual_income", "loan_to_income_ratio"]].head()
```

```python
df["risk_category"] = df["loan_paid_back"].map({1: "Low Risk", 0: "High Risk"})
```
